```python
class Emp:
    comp="HP"
    def __init__(self,name,sal):
        self.name=name
        self.sal=sal
        
    def printDetails(self):
        print(f"The Company of {self.name} is {self.comp} with salary ${self.sal}")

    @staticmethod
    def printTime():
        print(f"the time is now")
    @classmethod
    def printClassDetails(cls):
        print(f"the company is {cls.comp}")
e=Emp("Deepchandra",1200)
e.printDetails()
e.printTime()
e.printClassDetails()
```

```python
class Car:
    def __init__(self,brand , model):
        self.brand=brand
        self.model=model
    def displayInfo(self):
        return f"{self.brand} {self.model}"
# creating objectc(instance)

c1=Car("mahindra","thar")
print(c1.displayInfo())
```

```python
class BankAcount:
    def __init__(self,balance):
        self.balance=balance   #private attribute
    def get_balance(self): #Getter
        return self.balance
    def deposite(self,amount): #Setter
        if amount>0:
            self.balance+=amount
account=BankAcount(1000)
account.deposite(500)
print(account.get_balance())
```
