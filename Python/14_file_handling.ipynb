```python
# finding path 
import os 
print("current folder:",os.getcwd())
print("absulute path:",os.path.abspath("peep.txt"))
```

```python
with open("peep.txt", "r") as f:
    for line in f:
        print(line, end="")

print("Data written successfully!")


with open("peep.txt", "r") as f: 
    print(f.readlines()) 
print("Data written successfully!")
```
