```python
import json
```

```python
# lets write the function to load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data=load_data("data.json")
```

```python
# write a function to display users and their connections
def display_users(data):
    print("Users and their Connections:\n")
    for user in data["users"]:
        print(f"ID:{user['id']} - {user['name']} is freinds with : {user['freinds']} and liked pages are : {user['liked_pages']}")
        
    print("\nPages Information\n")
    for page in data["pages"]:
        print(f"{page['id']} : {page['name']}")

display_users(data)   
        
        
```
