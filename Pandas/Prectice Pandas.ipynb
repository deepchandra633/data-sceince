```python
import pandas as pd
df = pd.DataFrame({ 
    "Name": [ "Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Henry", "Ivy", "Jack", "Kevin", "Lily", "Mason", "Nina", "Oscar" ], 
    "Age": [ 22, 25, 23, 26, 24, 27, 21, 28, 22, 25, 29, 20, 24, 23, 26 ], 
    "Marks": [ 85, 78, 92, 88, 95, 67, 81, 90, 73, 84, 96, 76, 89, 91, 82 ], 
    "City": [ "Delhi", "Mumbai", "Pune", "Delhi", "Jaipur", "Mumbai", "Delhi", "Pune", "Bhopal", "Indore", "Jaipur", "Delhi", "Indore", "Bhopal", "Pune" ] }) 
print(df)
```

```python
# Show students whose Marks are greater than 80.
# Show students whose Age is less than 25.
print(df.query("Marks > 80"))
print("------------")
print(df.query("Age<25"))
```

```python
# Show students from Delhi.
# Show students whose Marks are between 80 and 90.
print(df.query("City == 'Delhi'"))
print("----------")
print(df[df["Marks"].between(80,90)])
```

```python
# Show students whose Age is greater than 22 and Marks are greater than 85.
# Show students whose City is not Mumbai.
print(df.query("Age > 22 and Marks > 85"))
print("----------")
print(df.query("City != 'Mumbai'"))
```

```python
# Show students whose Name starts with "D".
# Show students whose Name contains the letter "a".
print(df[df["Name"].str.startswith("D")])
print("----------")
print(df[df["Name"].str.contains("a")])
```

```python
df["Age Group"]=df["Age"].apply(lambda x : "Adult" if x>22 else "Minor")
```

```python
# Sort by Marks ascending.
# Sort by Marks descending.
# Sort by City then Marks.
# Find top 5 scorers.
# Find youngest 3 students.
print(df.sort_values("Marks"))
print("-----")
print(df.sort_values("Marks",ascending=False))
print("-----")
print(df.sort_values(["City","Marks"]))
print("-----")
print(df.sort_index())
print("-----")
print(df.nlargest(5,"Marks"))
print("-----")
print(df.nlargest(3,"Age"))
```

```python
import pandas as pd
df = pd.DataFrame({
    "Employee":["Amit","Priya","Rahul","Sara","Mohit","Neha","Karan","Pooja"], 
    "Department":["IT","HR","IT","HR","Sales","Sales","IT","HR"], 
    "Salary":[50000,45000,60000,48000,55000,58000,65000,47000], 
    "Sales":[100,80,120,90,150,140,130,85], 
    "Experience":[2,3,5,4,6,5,7,2] 
})
print(df)
```

```python
print(df.groupby("Department")) # ye kuch bhi output nahi dega 
```

```python
print(df.groupby("Department")["Salary"])#ye bhi kuch nahi dega output me 
```

```python
# bina aggregrate ke ham groupby use nahi ker sakte 
print(df.groupby("Department")["Salary"].mean())
```

```python
print(df.groupby("Department")["Salary"].sum())
print(df.groupby("Department")["Salary"].max())
print(df.groupby("Department")["Salary"].min())
print(df.groupby("Department")["Salary"].count())
```

```python
print(df.groupby(["Department","Sales"])["Salary"].sum())
```

```python
print(df.groupby("Department")["Salary"].agg(["mean","sum","max","min","count"]))
```

```python
res=df.groupby("Department")["Salary"].agg(
    score_mean="mean",
    score_sum="sum",
    score_max="max", 
)
print(res)
```

```python
groups=df.groupby("Department")
groups.get_group("HR")
groups.get_group("IT")
```

```python
import pandas as pd 
employees = pd.DataFrame({
    "Emp_ID":[101,102,103,104,105], 
    "Name":["Amit","Priya","Rahul","Sara","Neha"], 
    "Department":["IT","HR","IT","Sales","HR"] }) 
print(employees)
```

```python
salary = pd.DataFrame({
    "Emp_ID":[101,102,103,106], 
    "Salary":[50000,45000,60000,70000] }) 
print(salary)
```

```python
pd.merge(employees,salary,on="Emp_ID")
```

```python
pd.merge(employees,salary,on="Emp_ID",how="left")
```

```python
pd.merge(employees,salary,on="Emp_ID",how="right")
```

```python
pd.merge(employees,salary,on="Emp_ID",how="outer")
```
