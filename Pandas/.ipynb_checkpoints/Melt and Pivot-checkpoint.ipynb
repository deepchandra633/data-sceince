```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [85, 78, 92],
    'Science': [90, 82, 89],
    'English': [88, 85, 94]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
df2=df.melt(id_vars=["Name"],value_vars=['Math','Science','English'],var_name="Subject",value_name='Score').copy()
```

```python
df2
```

```python
df2.pivot(index="Name",columns="Subject",values="Score")
```

```python
result = df2.pivot(index="Name", columns="Subject", values="Score").reset_index()

result.columns.name = None

print(result)
```
