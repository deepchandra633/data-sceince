```python
import numpy as np
```

```python
arr = np.arange(1, 21)
```

```python
# Print only even numbers.
even_num=arr % 2 == 0
print(arr[even_num])
```

```python
# Print numbers greater than 10.
greater_num=arr > 10
print(arr[greater_num])
```

```python
# Find the sum of all even numbers.
even_sum=arr % 2 == 0
print("Sum of even numbers : ",np.sum(arr[even_sum]))
```

```python
# Find the mean of all odd numbers.
odd_mean=arr % 2 == 1
print("Mean of all odd numbers : ",np.mean(arr[odd_mean]))
```

```python
# Convert the array into a 4×5 matrix.
reshaped=np.reshape(arr,(4,5))
print(reshaped)
```

```python
# Print the last row.
# print(reshaped[3][:])
print(reshaped[3,:])

# Print the second column.
print(reshaped[:,1])
```

```python
# Find the maximum value in the matrix.
print(np.max(reshaped))
```

```python
# Find the index of value 15.
index = np.where(arr == 15)

print(index)

print(np.where(arr==15))
```

```python
# Reverse the array.
# print(arr[::-1])
print(reshaped[::-1])
```

```python
# Replace all numbers less than 15 with -1
reshaped[reshaped<15]=-1
print(reshaped)
```
