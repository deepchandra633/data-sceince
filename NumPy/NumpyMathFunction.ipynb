```python
import numpy as np
```

```python
arr=np.array([2,4,4,4,5,5,7,9])
```

```python
print(np.mean(arr))
```

```python
print(np.var(arr))
```

```python
print(np.std(arr))
```

```python
print(np.max(arr))
print(np.min(arr))
```

```python
print(np.prod(arr))
print(np.sum(arr))
```

```python
print(np.median(arr))
```

```python
print(np.mod(arr,2)) # gives reminder
```

```python
print(np.percentile(arr,30))  # ye btata he ki kitna hissa 50%,25,50,75 ke niche he ya uske baraber he 
```

```python
print(np.argmax(arr))
print(np.argmin(arr))
```

```python
print(np.unique(arr))
print(np.cumsum(arr))
print(np.cumprod(arr))
```

```python
print(np.log(arr)) # calculate on base e natural base bhi kahte he 
```

```python
print(np.exp(arr))
```

```python
print(np.add(arr,5))
print(np.multiply(arr,5))
print(np.subtract(arr,5))
print(np.power(arr,2))
print(np.divide(arr,2))
print(np.square(arr))
print(np.sqrt(arr))
print(np.cbrt(arr))
print(np.abs(arr))
```
