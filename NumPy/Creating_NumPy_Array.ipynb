```python
# indexing and slicing in numpy array

arr=np.array([1,2,3,4,5,6,78,90,8,6,12])
print(arr[2])
print(arr[-1])
print(arr[1:4])
print(arr[1:])
print(arr[:])
print(arr[:7])

# important slicing
print(arr[:3]) # isse to first three element aa jayge 
print(arr[::5]) # isse her position element aayga array sse jo ham denge 
```
