```python
# using python list
import numpy as np
import time

size=1_000_000
l1=list(range(size))
l2=list(range(size))
# l1=[1,2,3]
# l2=[4,5,6]
start=time.time()

add=[x+y for x,y in zip(l1,l2)]

end=time.time()
print(add[0:10])


print(end-start)
```

```python
# using NumPy Arrays
import numpy as np
import time

size=1_000_000
l1=np.array(list(range(size)))
l2=np.array(list(range(size)))

start=time.time()

add=l1+l2

end=time.time()
print(add[0:10])

print(end-start)
```

```python
# vecterization No More Loops

# python list(loop-based)
import time
list1=list(range(10))
start=time.time()
list_sqr=[x**2 for x in list1]
end=time.time()
print(end-start)
print(list_sqr)

# numpy(vexterized)
import numpy as np
arr1=np.array(list1)

start=time.time()
numpy_sqr=arr1 ** 2
end=time.time()
print(end-start)
print(numpy_sqr)

```

```python
# find the memory size of a NumPy array with 1 million elements 
import numpy as np

# Create an array with 1 million elements
arr = np.arange(1000000)

# Memory size in bytes
print("Memory Size (Bytes):", arr.nbytes)

# Memory size in KB
print("Memory Size (KB):", arr.nbytes / 1024)

# Memory size in MB
print("Memory Size (MB):", arr.nbytes / (1024 * 1024))
```
