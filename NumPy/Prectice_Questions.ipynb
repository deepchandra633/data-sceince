```python
# find the memory size of a NumPy array with 1 million elements 
import numpy as np

# Create an array with 1 million elements
arr = np.arange(1000000)

# Memory size in bytes
print("Memory Size (Bytes):", arr.nbytes)

# Memory size in KB
print("Memory Size (KB):", arr.nbytes / 1024)

# Memory size in MB
print("Memory Size (MB):", arr.nbytes / (1024 * 1024))
```

```python
# convert an array of float [1.1,2.2,3.3,4.4] int integer

arr = np.array([1.1, 2.2, 3.3, 4.4])

int_arr = arr.astype(int)   # astype(int) converts each element to an integer.

print("Original Array:", arr)
print("Integer Array:", int_arr)
```
